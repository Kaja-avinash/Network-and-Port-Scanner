```python
import os
import time
import sys
import socket
import random
import itertools
import requests
from tqdm import tqdm
from pyfiglet import Figlet
import pyqrcode
from barcode import EAN13
from barcode.writer import ImageWriter
from phonenumbers import geocoder, carrier, parse
from tabulate import tabulate


def font(text):
    return Figlet(font="slant").renderText(text)

def loading():
    for _ in tqdm(range(100), desc="LOADING...", ascii=False, ncols=75):
        time.sleep(0.01)
    print("LOADING DONE!")

def main_menu():
    os.system("cls" if os.name == "nt" else "clear")
    print(Figlet(font="5lineoblique").renderText("STRANGE"))
    print("##USE THIS AT YOUR OWN RISK\nMr.STRANGE IS NOT RESPONSIBLE FOR YOUR DEEDS")
    print("""
1- MY IP ADDRESS
2- PASSWORD GENERATOR
3- WORDLIST GENERATOR
4- BARCODE GENERATOR
5- QRCODE GENERATOR
6- PHONE NUMBER SCANNER
7- SUBDOMAIN SCANNER
    """)
    return int(input("ENTER YOUR CHOICE >>>>----------> "))

def get_ip():
    print(font("FIND MY IP"))
    loading()
    hostname = socket.gethostname()
    ip_address = socket.gethostbyname(hostname)
    print(f"YOUR DEVICE IS: {hostname}")
    print(f"YOUR IP ADDRESS IS: {ip_address}")
    input("PRESS ENTER TO EXIT")

def password_generator():
    print(font("PASSWORD GENERATOR"))
    loading()
    length = int(input("ENTER PASSWORD LENGTH >>>>----------> "))
    chars = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890@#&*(){}[]/?"
    password = "".join(random.sample(chars, length))
    print(f"GENERATED PASSWORD >>>>----------> {password}")
    input("PRESS ENTER TO EXIT")

def wordlist_generator():
    print(font("WORDLIST GENERATOR"))
    loading()
    chrs = input("ENTER CHARACTERS >>>>----------> ")
    min_len = int(input("MINIMUM LENGTH >>>>----------> "))
    max_len = int(input("MAXIMUM LENGTH >>>>----------> "))
    filename = input("FILENAME >>>>----------> ")
    total = 0
    with open(filename, 'w') as f:
        for i in range(min_len, max_len + 1):
            for item in itertools.product(chrs, repeat=i):
                f.write("".join(item) + "\n")
                total += 1
    print(f"WORDLIST GENERATED: {filename} ({total} words)")
    input("PRESS ENTER TO EXIT")

def barcode_generator():
    print(font("BARCODE GENERATOR"))
    loading()
    num = input("ENTER 12 DIGIT NUMBER >>>>----------> ")
    code = EAN13(num, writer=ImageWriter())
    code.save("bar_code")
    print("SAVED AS bar_code.png")
    input("PRESS ENTER TO EXIT")

def qrcode_generator():
    print(font("QRCODE GENERATOR"))
    loading()
    link = input("ENTER THE LINK >>>>----------> ")
    qr = pyqrcode.create(link)
    qr.png("myqr.png", scale=6)
    print("SAVED AS myqr.png")
    input("PRESS ENTER TO EXIT")

def phone_number_scanner():
    print(font("PHONE NUMBER SCANNER"))
    loading()
    number = input("ENTER NUMBER WITH COUNTRY CODE >>>>----------> ")
    num = parse(number)
    country = geocoder.description_for_number(num, 'en')
    provider = carrier.name_for_number(num, 'en')
    table = [["COUNTRY", "SUPPLIER"], [country, provider]]
    print(tabulate(table, headers="firstrow", tablefmt="github"))
    input("PRESS ENTER TO EXIT")

def subdomain_scanner():
    print(font("SUBDOMAIN SCANNER"))
    loading()
    domain = input("ENTER DOMAIN >>>>----------> ")
    try:
        with open("subdomains.txt") as f:
            subs = f.read().splitlines()
        for sub in subs:
            url = f"http://{sub}.{domain}"
            try:
                requests.get(url)
                print(f"[+] Found: {url}")
            except requests.ConnectionError:
                continue
    except FileNotFoundError:
        print("subdomains.txt not found!")
    input("PRESS ENTER TO EXIT")

def main():
    while True:
        choice = main_menu()
        if choice == 1:
            get_ip()
        elif choice == 2:
            password_generator()
        elif choice == 3:
            wordlist_generator()
        elif choice == 4:
            barcode_generator()
        elif choice == 5:
            qrcode_generator()
        elif choice == 6:
            phone_number_scanner()
        elif choice == 7:
            subdomain_scanner()
        else:
            print("INVALID CHOICE")
            input("PRESS ENTER TO CONTINUE")

if __name__ == '__main__':
    main()
```
